```python
import tensorflow as tf
import os

data_path = tf.keras.utils.get_file(
    fname='speech_commands_v0.02.tar.gz',
    origin='http://storage.googleapis.com/download.tensorflow.org/data/speech_commands_v0.02.tar.gz',
    extract=True
)

print("Downloaded file:", data_path)
print("Files in dataset folder:", os.listdir(os.path.dirname(data_path)))
```

```python
import os

print(os.listdir('/root/.keras/datasets/'))
```

```python
data_dir = '/root/.keras/datasets/'
classes = ["tree", "three", "two"]

for c in classes:
    folder = os.path.join(data_dir, c)
    print(c, len(os.listdir(folder)))
```
